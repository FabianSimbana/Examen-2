```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

def fprime(x):
    return 5*x**4 - 24*x**3 + 6*x**2 + 40*x - 27

x_vals = np.linspace(-2.5, 2.5, 400)


plt.plot(x_vals, polinomio(x_vals), label="Polinomio")
plt.grid(True)
plt.ylim(-0.1, 0.1)  
plt.xlim(0.75, 1.25)  
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico del Polinomio')
plt.legend()
plt.show()
```

```python
estimaciones_iniciales = np.linspace(0.9, 1.1, 6)

raices = []
for estimacion in estimaciones_iniciales:      
        raiz = optimize.newton(polinomio, estimacion, fprime)
        raices.append(raiz)

print("Las raíces del polinomio en el rango [0.9, 1.1] son:")
for raiz in raices:
    print(raiz)
```
